# ABSTRACT

People love skiing and selecting a good ski resort is quite a task. So in this project I have tried to make it easier for people to choose good ski resorts in California. This project gathers all information about ski resort including ticket price, ski conditions, weather of its city, ratings and so on from Google and Instagram. Then I created a database with all this information of the resorts. When customer insert their requirement on our database, the best recommendations will be provided to them.

# Data Resource 

WIKIPEDIA: The ski resorts information in California was obtained from the wikipedia page: https://en.wikipedia.org/wiki/Comparison_of_California_ski_resorts. The data contains resort names, nearest cities, skiing conditions etc.



GOOGLE: The address of the ski resorts, ratings, google id etc. was obtained from Google using Google Maps Web API  




INSTAGRAM: The hashtags, captions and likes for these ski resorts was obtained from Instagram using Instagram Web API
